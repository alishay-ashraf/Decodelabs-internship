```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris benchmark dataset
iris_data = load_iris()

# Convert to a DataFrame for clean visualization
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['target'] = iris_data.target

print("--- Dataset Shape ---")
print(f"Samples: {df.shape[0]}, Dimensions: {df.shape[1] - 1}\n")

print("--- First 5 Rows of Raw Data ---")
print(df.head())
```

```python
from sklearn.model_selection import train_test_split

# Separate features (X) and labels (y)
X = df.drop(columns=['target'])
y = df['target']

# Split into 80% Training and 20% Testing (with random shuffling)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=True)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit on training data and transform both sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Features scaled successfully! Mean is now ~0 and Variance is ~1.")
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the KNN classifier (using K=3 neighbors as a start)
knn_model = KNeighborsClassifier(n_neighbors=3)

# Train the model using pure algorithmic logic
knn_model.fit(X_train_scaled, y_train)

# Make predictions on the validation test set
y_pred = knn_model.predict(X_test_scaled)
print("Model training complete. Predictions generated!")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score

# Generate the Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
# Calculate macro F1-Score
macro_f1 = f1_score(y_test, y_pred, average='macro')

print("--- Confusion Matrix ---")
print(conf_matrix)
print(f"\nMacro F1-Score: {macro_f1:.4f}\n")

print("--- Detailed Classification Report ---")
print(classification_report(y_test, y_pred, target_names=iris_data.target_names))
```
